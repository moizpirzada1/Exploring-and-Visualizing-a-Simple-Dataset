```python
# 1. Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2. Load the Iris Dataset
df = sns.load_dataset('iris')
```

```python
# 3. Inspect the Dataset
# -----------------------------
print("Shape of dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nFirst five rows:\n", df.head())
```

```python
# Summary information
print("\nDataset Info:")
print(df.info())
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
# 4. Data Visualization
# (a) Scatter Plot - Sepal Length vs Petal Length
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='deep')
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.show()
```

```python
# (b) Histograms - Distribution of all numeric features
df.hist(edgecolor='black', figsize=(10, 8))
plt.suptitle("Histograms of Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# (c) Boxplots - Identify Outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient="h", palette="Set2")
plt.title("Box Plots for Outlier Detection")
plt.show()
```
